```python
import json
import numpy as np
import pandas as pd
import altair as alt

# Hilfsfunktion (muss hier nochmal stehen — anderes Notebook, neue Umgebung)
def hist_to_df(train_hist, val_hist, value_name):
    rows  = [{"Epoch": ep, value_name: v, "Type": "train"} for ep, v in train_hist.items()]
    rows += [{"Epoch": ep, value_name: v, "Type": "validation"} for ep, v in val_hist.items()]
    df = pd.DataFrame(rows)
    return df[np.isfinite(df[value_name])]

# Run-ID hier anpassen
run_id = 2
with open(f"../results/run_{run_id}.json") as f:
    h = json.load(f)

train_loss = dict(enumerate(h["train_loss"]))
val_loss   = dict(enumerate(h["val_loss"]))
train_acc  = dict(enumerate(h["train_acc"]))
val_acc    = dict(enumerate(h["val_acc"]))

loss_df = hist_to_df(train_loss, val_loss, "Loss")
acc_df  = hist_to_df(train_acc,  val_acc,  "Accuracy")

loss_chart = alt.Chart(loss_df).mark_line().encode(
    x="Epoch", y="Loss", color="Type").properties(title=f"Loss – Run {run_id}", width=350)
acc_chart = alt.Chart(acc_df).mark_line().encode(
    x="Epoch", y="Accuracy", color="Type").properties(title=f"Accuracy – Run {run_id}", width=350)

loss_chart | acc_chart
```
